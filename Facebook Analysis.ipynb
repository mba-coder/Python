```python
predictor_vars=['Gender ',
 'Emp',
 'Profile',
 'Cover',
 'Orientation',
 'Relationship',
 'Posts',
 'Replies',]
```

```python
for var in predictor_vars:
    X = Facebook[var]  # Predictor variable
    y = Facebook['Friends']  # Response variable

    # Add a constant to the predictor variable for the intercept term
    X = sm.add_constant(X)

    # Fit the linear regression model
    model = sm.OLS(y, X).fit()

    # Print the summary of the regression model
    print(f"Regression Analysis for {var}:")
    print(model.summary())
    
```

```python
import statsmodels.api as sm

# Fit the linear regression model using the formula method
model = sm.OLS.from_formula(formula='Friends ~ Tags', data=Facebook).fit()

# Print the summary of the regression model
print(model.summary())
```
